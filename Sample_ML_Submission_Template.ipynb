# **Project Summary -**

The project focuses on analyzing customer interaction data to uncover patterns, identify bottlenecks, and suggest actionable strategies for improving customer satisfaction and operational efficiency. The dataset contains over 85,000 customer interactions across different channels, categories, and time periods, with detailed information about agents, supervisors, products, handling time, and customer satisfaction scores (CSAT).

---

## Data Understanding and Cleaning

The dataset consisted of 20 columns, including both categorical and numerical features. Some variables such as *Customer Remarks, Order ID, Order Date/Time, Customer City,* and *Connected Handling Time* contained a high percentage of missing values. After careful assessment, these were handled through appropriate wrangling steps, such as dropping irrelevant features, imputing values where possible, and engineering new ones like *Order Month* and *Order Day*. Duplicates were checked and found to be zero, ensuring data quality.

Exploratory data analysis (EDA) revealed that most numerical variables were skewed. *Item Price* had extreme outliers with a few very expensive items, while *Connected Handling Time* had long tails, indicating some cases that took much longer than average. CSAT scores, on the other hand, were tightly clustered between 4 and 5, showing that overall customer satisfaction was stable but still sensitive to certain operational challenges.

---

## Key Visualizations and Insights

Several visualizations were created to tell a story from the data.

* **Channel Analysis** showed that inbound calls dominate interactions, followed by outcalls and email. While calls are the main touchpoint, email was found to have greater variation in CSAT, suggesting room for improvement.

* **Category and Sub-category Distribution** highlighted that a majority of customer complaints came from returns, order-related queries, and delays. Pareto analysis confirmed that a small set of issues accounts for nearly 80% of all customer concerns.

* **CSAT by Channel and Category** revealed that while most channels maintain fairly high satisfaction, email interactions had more outliers, and some categories such as refunds and cancellations showed slightly lower CSAT, signaling the need for process improvement.

* **Agent and Supervisor Analysis** exposed a wide gap between top and bottom performers. Some agents and supervisors consistently achieved near-perfect scores, while others lagged behind. This indicates that individual performance has a significant effect on customer satisfaction.

* **Tenure Analysis** suggested that CSAT improves with agent experience, but dips after 90 days. This points to onboarding challenges for new agents and motivational issues for long-term employees.

* **Handling Time Distribution** showed that most cases fall within 200–600 seconds, but a few extreme cases push times beyond 1000 seconds. This highlights inefficiencies that need process standardization.

* **Seasonality and Weekly Trends** revealed clear peaks in July and August, as well as on weekends, especially Fridays. This confirms the need for flexible resource planning during high-demand periods.

* **Correlation and Pair Plot Analysis** confirmed that expensive items take longer to resolve (moderate correlation of 0.65 between price and handling time), but handling time itself has little effect on CSAT, showing that customers value quality of resolution more than speed.

---

## Business Recommendations

The analysis suggests several strategies to achieve business objectives:

1. Focus on the top issue categories like returns, order delays, and pickups to remove the bulk of customer pain points.
2. Improve email support with automation and templates to reduce call center load and ensure consistent quality.
3. Use top-performing agents and supervisors as role models for training and mentoring underperforming staff.
4. Strengthen onboarding training for new agents while introducing long-term incentives to keep experienced staff motivated.
5. Standardize processes and provide better knowledge bases for complex cases to reduce extreme handling times.
6. Adjust staffing levels to account for seasonal spikes (July–August) and weekly peaks (Friday–Sunday).

## Conclusion

This project demonstrates that customer satisfaction is less about speed or product value and more about resolving the right issues with quality and consistency. By focusing on the biggest pain points, leveraging high-performing employees as benchmarks, and planning operations around demand peaks, the business can significantly improve both efficiency and customer experience. These insights provide a clear roadmap for achieving better customer trust, loyalty, and operational effectiveness.

#### Chart - 3

````python
# Step 1: Get frequency of each sub-category
subcat_counts = df['Sub-category'].value_counts()

# Step 2: Calculate cumulative percentage
cum_percentage = subcat_counts.cumsum() / subcat_counts.sum() * 100

# Step 3: Plot Pareto chart
fig, ax1 = plt.subplots(figsize=(12,6))

# Bar plot (counts)
ax1.bar(subcat_counts.index, subcat_counts.values, color="skyblue")
ax1.set_xlabel("Sub-category", fontsize=12)
ax1.set_ylabel("Count", color="blue", fontsize=12)
ax1.tick_params(axis="y", labelcolor="blue")
plt.xticks(rotation=90)

# Line plot (cumulative %)
ax2 = ax1.twinx()
ax2.plot(subcat_counts.index, cum_percentage, color="red", marker="D", ms=4)
ax2.set_ylabel("Cumulative %", color="red", fontsize=12)
ax2.tick_params(axis="y", labelcolor="red")
ax2.axhline(80, color="green", linestyle="--", linewidth=1)  # 80% reference line

# Title
plt.title("Pareto Analysis of Customer Interactions by Sub-category", fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A Pareto chart was chosen because Sub-category has many unique values. Instead of showing all counts equally, Pareto helps identify the few subcategories that contribute to the majority of customer issues (80/20 rule). This makes the visualization focused and more business-relevant.

##### 2. What is/are the insight(s) found from the chart?

- A small set of subcategories drives most customer issues. For example:

- The cumulative line shows that about the top 6–8 subcategories account for ~80% of all issues.

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothesis 1: CSAT scores differ across service channels.

- Alternate (H₁): Longer handling times reduce CSAT scores.

Hypothesis 3: CSAT scores differ across issue categories.

- Null (H₀): There is no significant difference in CSAT scores across categories (Returns, Order Related, Refund, etc.).

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- Null (H₀): There is no significant difference in CSAT scores between channels (Inbound, Outcall, Email).

- Alternate (H₁): CSAT scores vary significantly between channels.

- Test to use: One-Way ANOVA (since CSAT Score is numeric and channel_name is categorical with >2 groups).

#### 2. Perform an appropriate statistical test.

````python
alpha = 0.05  # significance level


anova_channels = stats.f_oneway(
    df[df['channel_name'] == 'Inbound']['CSAT Score'],
    df[df['channel_name'] == 'Outcall']['CSAT Score'],
    df[df['channel_name'] == 'Email']['CSAT Score']
)

print("Hypothesis 1 (CSAT vs Channel)")
print("F-statistic:", anova_channels.statistic, "p-value:", anova_channels.pvalue)
if anova_channels.pvalue < alpha:
    print("Reject Null Hypothesis: CSAT differs significantly across channels.\n")
else:
    print("Fail to Reject Null Hypothesis: No significant difference across channels.\n")
````

### 2. Handling Outliers

````python
# Function to cap outliers using IQR
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series.clip(lower, upper)

# Apply to Item_price
df['Item_price'] = cap_outliers(df['Item_price'])

# Apply to connected_handling_time (if kept)
if 'connected_handling_time' in df.columns:
    df['connected_handling_time'] = cap_outliers(df['connected_handling_time'])
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outlier Treatment Techniques Used

Technique Applied:

- I used the Interquartile Range (IQR) method with capping (winsorization).

- For each numeric variable (Item_price and Connected Handling Time), the 25th percentile (Q1) and 75th percentile (Q3) were calculated.

- The Interquartile Range (IQR = Q3 – Q1) was used to define acceptable bounds:
  LowerBound=Q1–1.5×IQR,UpperBound=Q3+1.5×IQR

- Any value below the lower bound was capped to the lower limit, and values above the upper bound were capped to the upper limit.

Why I used this technique:

- Outliers can skew model training, especially when numeric values vary widely.

- Instead of removing rows (which would reduce data size), capping ensures the extreme values are controlled without losing records.

- This method keeps the dataset balanced and reduces the undue influence of extreme cases, while still preserving meaningful business scenarios.

Outcome:

- Both Item_price and Connected Handling Time distributions became more stable.

- The data is now less skewed and more suitable for modeling.

#### 8. Text Normalization

````python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def normalize_text(text):
    if not isinstance(text, str):
        return text
    
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    
    # 3. Remove words with digits
    text = re.sub(r'\w*\d\w*', '', text)
    
    # 4. Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # 5. Remove stopwords
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    
    # 6. Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # 7. Remove extra whitespaces and join back
    text = ' '.join(tokens)
    
    return text

# Apply normalization to Customer Remarks
df['Customer Remarks Cleaned'] = df['Customer Remarks'].apply(normalize_text)

# Example
sample_text = "Wow!!! This product was running 2days late, but it's AMAZING! Check http://example.com"
print("Before:", sample_text)
print("After :", normalize_text(sample_text))
````

##### Which text normalization technique have you used and why?

In this project, the following text normalization techniques were applied to the Customer Remarks column:

1. Lowercasing

- Converted all text to lowercase.

-  Why: Ensures consistency, so words like “Good”, “good”, “GOOD” are treated as the same token.

2. Removing Punctuation & Special Characters

- Stripped symbols such as ! , . ? @ #.

- Why: These don’t add semantic meaning in sentiment analysis or clustering and only add noise.

3. Removing URLs and Words with Digits

- Removed links (e.g., http://example.com
) and alphanumeric codes (e.g., order123).

-  Why: These are not useful for text analysis and increase vocabulary size unnecessarily.

4. Removing Stopwords

- Eliminated common words like the, is, and, but.

- Why: Stopwords don’t carry significant meaning and removing them makes the text more meaningful for NLP models.

5. Lemmatization

- Reduced words to their base form (running → run, better → good).

- Why: Decreases vocabulary size and groups word variations under the same root for better model generalization.

6. Whitespace Normalization

- Removed extra spaces and trimmed the text.

- Why: Ensures clean, uniform input for tokenization and vectorization.

Outcome

These normalization techniques transformed raw, inconsistent customer remarks into a clean, standardized form that is easier for NLP models to process.

#### 3. Explain each evaluation metric's indication towards business and the business impact pf the ML model used.

##  Evaluation Metrics and Their Business Indications

### 1. **Accuracy**

* **What it means**: Percentage of correctly predicted CSAT scores out of total predictions.
* **Business view**:

  * High accuracy = the model is generally reliable in predicting customer satisfaction levels.
  * Low accuracy = model is missing many signals, and decisions based on it (like agent training or process improvements) may not be fully effective.
* **Impact**: Since customer satisfaction directly affects retention and revenue, **higher accuracy ensures reliable insight** into where customers are happy or dissatisfied.

---

### 2. **Precision**

* **What it means**: Out of all the times the model predicted a particular CSAT class (say, dissatisfied customers), how many were correct.
* **Business view**:

  * High precision in predicting **dissatisfied customers (CSAT=1,2)** = fewer false alarms → the company doesn’t waste resources investigating customers who are actually satisfied.
  * Low precision = business teams might **overreact to wrong signals**, wasting support or recovery efforts on the wrong customers.
* **Impact**: Precision reduces **unnecessary customer outreach**, optimizing cost of corrective actions like refunds, credits, or agent re-training.

---

### 3. **Recall (Sensitivity)**

* **What it means**: Out of all actual cases of a class (e.g., truly dissatisfied customers), how many the model successfully identified.
* **Business view**:

  * High recall for **dissatisfied customers** = fewer unhappy customers are missed.
  * Low recall = risk of **missing real pain points**, meaning dissatisfied customers slip through unnoticed and may churn.
* **Impact**: Recall is **crucial for customer retention** — missing unhappy customers directly leads to lower repeat business and higher churn.

---

### 4. **F1-Score**

* **What it means**: The harmonic mean of precision and recall — balances the trade-off.
* **Business view**:

  * High F1-score means the business is **not only identifying dissatisfied customers but also doing so accurately**.
  * Low F1-score = imbalance (either too many false alarms or missing too many unhappy customers).
* **Impact**: F1 is the **most business-relevant metric** when dissatisfaction is rare but costly (since both false positives and false negatives have a cost).

---

### 5. **Confusion Matrix**

* **What it means**: Breaks down predictions into true positives, false positives, false negatives, and true negatives for each CSAT class.
* **Business view**:

  * Shows where the business is doing well (e.g., correctly classifying highly satisfied customers) and where it’s failing (e.g., confusing dissatisfied customers with satisfied ones).
* **Impact**: Helps identify **specific gaps** (e.g., if the model always predicts “satisfied” but misses “dissatisfied,” the company risks **overestimating satisfaction**).

---

##### Which hyperparameter optimization technique have you used and why?

Hyperparameter Optimization Technique Used:

I used RandomizedSearchCV instead of GridSearchCV.

Why RandomizedSearchCV?

1. GridSearchCV tries all possible combinations of parameters in the grid.

    - Example: If you have 6 parameters with 5 options each → 5⁶ = 15,625 combinations 🚨 (very slow).

    - With your dataset size (~85k rows), this becomes computationally heavy and caused errors (BrokenProcessPool).

2. RandomizedSearchCV samples a fixed number of random combinations from the parameter grid.

    - Example: If you set n_iter=20, it will only test 20 random parameter sets, instead of thousands.

    - Much faster and still effective at finding good hyperparameters.

3. It avoids pickling errors and parallelization crashes that happened in your GridSearchCV runs.

    - By using n_jobs=1, RandomizedSearchCV runs reliably even with complex models like XGBoost.

4. Business reason: Since the goal is to improve CSAT prediction efficiently, RandomizedSearchCV gives a good tradeoff between model performance and time/compute cost, making it practical for business use.

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

## ✅ Evaluation Metrics Considered

### 1. **Accuracy**

* **What it shows**: Proportion of total correct predictions.
* **Why it matters**: Gives a quick overview of model performance.
* **Business impact**:

  * Alone, accuracy can be misleading because the dataset is imbalanced (class 5 dominates).
  * Example: If 70% of customers give a CSAT score of 5, a “dummy model” always predicting 5 would give 70% accuracy but **miss all dissatisfied customers**.
* ✅ Considered, but not the primary business metric.

---

### 2. **Precision**

* **What it shows**: Out of all customers predicted to belong to a class, how many are actually correct.
* **Why it matters**: High precision for low-CSAT classes means we can trust the model when it says a customer is dissatisfied.
* **Business impact**:

  * If precision is low, the business might **waste resources chasing customers who are actually satisfied**.
  * So, improving precision reduces **false alarms in customer follow-ups**.

---

### 3. **Recall**

* **What it shows**: Out of all actual customers in a class, how many the model correctly identified.
* **Why it matters**: High recall for low CSAT scores (1–3) means fewer **dissatisfied customers slipping through the cracks**.
* **Business impact**:

  * Missing out on unhappy customers = **missed opportunity to improve retention**.
  * So, recall is critical for identifying **at-risk customers**.

---

### 4. **F1-Score**

* **What it shows**: Harmonic mean of precision and recall.
* **Why it matters**: Balances both precision and recall.
* **Business impact**:

  * Ensures that both aspects — **not wasting resources (precision)** and **not missing dissatisfied customers (recall)** — are optimized together.
  * Especially important for **minority classes (1–4 CSAT scores)**.

---

### 5. **Macro vs Weighted Averages**

* **Macro average**: Treats all classes equally, useful to see if the model is learning minority dissatisfaction classes.
* **Weighted average**: Weighs by class size, useful to see overall performance considering imbalanced data.
* **Business impact**:

  * Macro: Helps ensure small but **critical dissatisfied groups** are not ignored.
  * Weighted: Reflects real-world business performance.

---

##  Final Business Justification

For **positive business impact**, I focused mainly on:

* **Recall** for dissatisfaction classes (1–3 CSAT) → because missing unhappy customers hurts retention.
* **Precision** for dissatisfaction classes → because false alarms waste customer service resources.
* **F1-score (Macro & Weighted)** → because it balances the above two and ensures both majority and minority classes are addressed.

Accuracy was checked, but it’s not reliable for imbalanced datasets, so the business should **prioritize Recall + Precision (F1)** instead.

---

In short:
I considered **Recall, Precision, and F1-score** as the most meaningful metrics for business because they directly impact **customer retention, resource allocation, and service quality improvements**.
